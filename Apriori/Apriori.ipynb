```python
import pandas as pd

from apyori import apriori
```

```python
csv_dir_treinamento = "/content/sample_data/MercadoSim.csv"
data = pd.read_csv(csv_dir_treinamento, delimiter=";")
```

```python
transactions = list()
for i in range(len(data)):
  transactions.append([str(data.values[i, j]) for j in range(data.shape[1])])
```

```python
transactions
```

```python
# Execução do algoritmo Apriori
rules = apriori(transactions, min_support=0.003, min_confidence=0.008)
results = list(rules)
print(results)
```

```python
for item in results:
  if not 'nan' in item[0]: 
    print("Rule: " + str(item[0]))
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
  else: continue
```
